```python
import math
from math import *
from decimal import *
import matplotlib.pyplot as plt
```

```python
def calcInside(degree, sides):
  degree = degree / 2
  x = 2 * math.sin(math.radians(degree))
  return Decimal((sides * x) / 2)
```

```python
calcInside(60, 6)
```

```python
def calcOutside(degree, sides):
  degree = degree / 2
  x = 2 * math.tan(math.radians(degree))
  return Decimal((sides * x) / 2)
```

```python
calcOutside(60, 6)
```

```python
calcOutside(60, 6)
```

```python
def calcPi(n, pSide, radiues=1, degree=60.0):
  for i in range(n):
    a = calcInside(degree, pSide)
    b = calcOutside(degree, pSide)

    print(a, '<', b)
    degree = degree / 2
    pSide = pSide * 2
```
